```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x, y, iterations=1000, learning_rate=0.001):
    m_curr = 0.0
    b_curr = 0.0
    n = len(x)

    # scatter only once
    plt.figure()
    plt.scatter(x, y, c="red", marker="+", linewidths=2)

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr

        # compute
        cost = (1/n) * np.sum((y - y_predicted)**2)
        md = -(2/n) * np.sum(x * (y - y_predicted))
        bd = -(2/n) * np.sum(y - y_predicted)

        m_curr -= learning_rate * md
        b_curr -= learning_rate * bd

        # print(m_curr, b_curr)
        print(f"iter {i}: m = {m_curr:.4f}, b = {b_curr:.4f}, cost = {cost}")

        # plot only occasionally (not 10000 times)
        if i % 2 == 0 or i == iterations - 1:
            plt.plot(x, y_predicted)

    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()

x = np.array([1,2,3,4,5], dtype=float)
y = np.array([5,7,9,11,13], dtype=float)
gradient_descent(x, y, iterations=1000, learning_rate=0.05)
```
