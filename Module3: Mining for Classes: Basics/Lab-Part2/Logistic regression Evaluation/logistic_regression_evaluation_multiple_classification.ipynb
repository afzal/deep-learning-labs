<h2 style='color:blue' align="center">Logistic Regression: Multiple Classification</h2>

```python
from sklearn.datasets import load_digits
%matplotlib inline
import matplotlib.pyplot as plt
digits = load_digits()
```

```python
type(digits)
```

```python
plt.gray()
for i in range(2):
    plt.matshow(digits.images[i])
```

```python
dir(digits)
```

```python
digits.data
```

```python
digits.data[0]
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data,digits.target, test_size=0.2)
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

<h4 style='color:purple'>Measure accuracy of our model</h4>

```python
model.score(X_test, y_test)
```

<h4 style='color:purple'>Confusion Matrix</h4>

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
# Plot Confusion Matrix
# This draws a nice figure.

disp = ConfusionMatrixDisplay(confusion_matrix=cm)

plt.figure(figsize=(6, 5))
disp.plot(values_format="d")   # "d" means show integers
plt.title("Confusion Matrix")
plt.show()
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
print(classification_report(y_test, y_predicted))
```

```python
#In the example above, classes are balances, if classes are imbalanced then acc -> balanced_accuracy
from sklearn.metrics import balanced_accuracy_score

bal_acc = balanced_accuracy_score(y_test, y_predicted)
print("Balanced Accuracy:", bal_acc)
```

```python
##In the example above, classes are balances, if classes are imbalanced then ER -> BER
ber = 1 - balanced_accuracy_score(y_test, y_predicted)
print("BER:", ber)
```
