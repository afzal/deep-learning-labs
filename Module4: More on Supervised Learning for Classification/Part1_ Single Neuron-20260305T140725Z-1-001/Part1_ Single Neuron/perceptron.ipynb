```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification

# Create n-class dataset (2 features so we can plot)
X, y = make_classification(
    n_samples=150,
    n_features=2,
    n_classes=4,              # <-- n# of classes
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=1,  # one blob per class (still easy)
    class_sep=2.2,
    random_state=7
)

# Scatter plot
plt.figure(figsize=(6,4))

plt.scatter(X[y==0,0], X[y==0,1], marker='o', label='Class 0')
plt.scatter(X[y==1,0], X[y==1,1], marker='x', label='Class 1')
plt.scatter(X[y==2,0], X[y==2,1], marker='^', label='Class 2')
plt.scatter(X[y==3,0], X[y==3,1], marker='s', label='Class 3')

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("3-Class Synthetic Dataset")
plt.legend()
plt.show()
```

- **1.0** → medium  
- **1.5** → easy (what we used)  
- **2–3** → very easy

```python
import numpy as np
from sklearn.datasets import make_classification

# ---------------------------------------------------------
# Generate a simple linearly separable dataset (2 features) - it generates synthetic (artificial) data.
# ---------------------------------------------------------
X, y = make_classification(
    n_samples=100,
    n_features=2,
    n_classes=2,
    n_informative=2, #BOTH features are useful.
    n_redundant=0, #no useless duplicated features.
    n_repeated=0, #no repeated columns
    n_clusters_per_class=1, #Each class (label −1 or +1) can be made of one or more clusters (blobs).
    class_sep=1.5, #Controls distance between classes, Larger value → easier problem; Smaller value → overlapping points.
    random_state=10 #it’s reproducible randomness.
)
```

```python
np.unique(y)
```

```python

# ---------------------------------------------------------
#  Convert labels from {0,1} to {-1,+1}
# ---------------------------------------------------------
y[y == 0] = -1
```

```python
np.unique(y)
```

```python
# Scatter plot
plt.figure(figsize=(6,4))

plt.scatter(X[y==-1,0], X[y==-1,1], marker='x', label='Class -1')
plt.scatter(X[y==+1,0], X[y==+1,1], marker='o', label='Class +1')

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Linearly Separable Dataset (2 Features)")
plt.legend()
plt.show()
```

```python
print(X[:5])
print(y[:5])
```

```python

# ---------------------------------------------------------
#  Add a column of ones to X (bias trick)
#    Now each sample becomes [1, x1, x2]
# ---------------------------------------------------------
X = np.append(np.ones((len(X), 1)), X, axis=1)
```

```python
print(X[:5])
print(y[:5])
```

```python

# ---------------------------------------------------------
#   Example neuron output using sign( w^T x )
#    (w includes the bias as its first element)
# ---------------------------------------------------------
w = np.array([0.2, 1.0, -0.6])   # [b, w1, w2]  (example weights)

n = 0  # choose a sample index
# Step 1: compute weighted sum (Σ = w^T x)
z = w.T.dot(X[n])

# Step 2: apply sign activation
y_hat_n = np.sign(z)

print("X[n] =", X[n])
print("Y[n] =", y[n])
print("Weighted sum z =", z)
print("Prediction =", y_hat_n)
```

```python

# If you want predictions for all samples:
y_hat_all = np.sign(X.dot(w))
print("First 10 predictions:", y_hat_all[:10])
```
